```python
#Create a simple Pandas Series from a list
a = [8, 23, 21]
myvar = pd.Series(a)
print(myvar)
print('\n')
print(myvar[0])
print(myvar[2])
```

```python
#changing index
a = [1, 7, 2]
myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
print('\n')
print(myvar["y"])
```

```python
#key/value object series
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
print('\n')
#printing selected keys
myvar = pd.Series(calories, index = ["day1", "day3"])
print(myvar)
```

```python
#Framing data of two series
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
df = pd.DataFrame(data)
print(df)
print('\n')
#refer particular location 
print(df.loc[0])
print('\n')
print(df.loc[[0, 1]])
print('\n')
print(df.loc[[0, 2]])
print('\n')
#changing index
df = pd.DataFrame(data, index = ["day1", "day2", "day3"])
print(df) 
#refering to particular location
print(df.loc["day2"])
```
